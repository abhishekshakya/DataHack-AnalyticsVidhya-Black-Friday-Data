```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#getting the data and cleaning
data = pd.read_csv('test.csv')
data.head()
```

```python
data.info()
```

```python
#assuming cutomer hasnot purschased product of that category thats why cell is left blank
# replacing NaN with 0
data.fillna(value=0,inplace=True)
data.head()
```

```python
#dropping userID and ProductID
data.drop(['User_ID','Product_ID'],axis=1,inplace=True)
data.head()
```

```python
#making Gender as binary
data['Gender'].replace(to_replace=['F','M'],value=[0,1],inplace=True)
data.head()
```

```python
#now lets Hot-Encode these
new_age = pd.get_dummies(data['Age'],prefix='Age',drop_first=True)
new_age.head()
```

```python
data.drop(['Age'],axis=1,inplace=True)
data = pd.concat(objs=[data,new_age],axis=1)
data.head()
```

```python
#lets do the same for City-Category
new_city = pd.get_dummies(data['City_Category'],prefix='city',drop_first=True)
new_city.head()
```

```python
data.drop(['City_Category'],axis=1,inplace=True)
data = pd.concat(objs=[data,new_city],axis=1)
data.head()
```

```python
#lets Drop Stay of people in city(may be that would'nt affect the purchase price)
data.drop(['Stay_In_Current_City_Years'],axis=1,inplace=True)
data.head()
```
