```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#getting the data and cleaning
data = pd.read_csv('train.csv')
data.head()
```

```python
data.info()
```

```python
#Product2 and PRoduct3 have some null values
```

```python
print(f"product2 = {data['Product_Category_2'].isna().sum()} and product3 = {data['Product_Category_3'].isna().sum()} null")
```

```python
#assuming cutomer hasnot purschased product of that category thats why cell is left blank
# replacing NaN with 0
data.fillna(value=0,inplace=True)
data.head()
```

```python
#dropping userID and ProductID
data.drop(['User_ID','Product_ID'],axis=1,inplace=True)
data.head()
```

```python
#making Gender as binary
data['Gender'].replace(to_replace=['F','M'],value=[0,1],inplace=True)
data.head()
```

```python
#Lets find number of categories in Age
data['Age'].unique()
```

```python
#now lets Hot-Encode these
new_age = pd.get_dummies(data['Age'],prefix='Age',drop_first=True)
new_age.head()
```

```python
data.drop(['Age'],axis=1,inplace=True)
data = pd.concat(objs=[data,new_age],axis=1)
data.head()
```

```python
#lets do the same for City-Category
new_city = pd.get_dummies(data['City_Category'],prefix='city',drop_first=True)
new_city.head()
```

```python
data.drop(['City_Category'],axis=1,inplace=True)
data = pd.concat(objs=[data,new_city],axis=1)
data.head()
```

```python
#lets Drop Stay of people in city(may be that would'nt affect the purchase price)
data.drop(['Stay_In_Current_City_Years'],axis=1,inplace=True)
data.head()
```

```python
#split the data into training and test data
X = data['Purchase'].values
data.drop(['Purchase'],inplace=True,axis=1)
data.head()
```

```python
inp = data.iloc[:,:].values
inp
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(inp,X,test_size = 0.2)
```

```python
#lets apply ML Models
#1Decision Tree
```

```python
from sklearn.tree import DecisionTreeRegressor
classifier = DecisionTreeRegressor()
classifier.fit(x_train,y_train)
```

```python
y_pred = classifier.predict(x_test)
```

```python
from sklearn.metrics import mean_squared_error
err = mean_squared_error(y_test,y_pred)
```

```python
np.sqrt(err)
```

```python
from sklearn.ensemble import RandomForestRegressor
reg = RandomForestRegressor()
reg.fit(x_train,y_train)
```

```python
y_pred2 = reg.predict(x_test)
```

```python
err2 = mean_squared_error(y_test,y_pred2)
np.sqrt(err2)
```
